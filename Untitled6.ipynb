```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)
```

```python
credit_data = pd.read_csv('/content/german.csv',sep=';')
```

```python
credit_data.to_sql('credit_data',engine)
```

```python
# Create an in-memory SQLite database
df = credit_data
import sqlite3
conn = sqlite3.connect(':memory:')
df.to_sql('credit_data', conn, index=False, if_exists='replace')
def run_query(query):
    return pd.read_sql(query, conn)
run_query("SELECT * FROM credit_data LIMIT 5")
```

```python
credibilty_record = run_query("""
    SELECT Creditability, COUNT(*) as count
    FROM credit_data
    GROUP BY Creditability
""")
import matplotlib.pyplot as plt
labels = ['Not Credible', 'Credible']
colors = ['red', 'green']
credibilty_record.plot(kind='pie', x='Creditability', y='count', autopct='%1.1f%%',labels = labels,colors = colors,startangle=90,shadow=True,explode=(0,0.1),legend = False)
plt.title('Creditability Record')
plt.show()
```

```python
run_query("""
    SELECT Account_Balance,
           SUM(CASE WHEN Creditability = 1 THEN 1 ELSE 0 END) as good_credit,
           SUM(CASE WHEN Creditability = 0 THEN 1 ELSE 0 END) as bad_credit,
           ROUND(SUM(CASE WHEN Creditability = 0 THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) as bad_rate
    FROM credit_data
    GROUP BY Account_Balance
    ORDER BY Account_Balance
""")
```

```python
durationwise_credibilty = run_query("""
    SELECT
        CASE
            WHEN Duration_of_Credit_monthly <= 12 THEN '0-12 months'
            WHEN Duration_of_Credit_monthly <= 24 THEN '13-24 months'
            WHEN Duration_of_Credit_monthly <= 36 THEN '25-36 months'
            ELSE '37+ months'
        END as duration_range,
        COUNT(*) as total,
        SUM(Creditability) as good_credit,
        COUNT(*) - SUM(Creditability) as bad_credit,
        ROUND((COUNT(*) - SUM(Creditability)) * 100.0 / COUNT(*), 2) as bad_rate
    FROM credit_data
    GROUP BY duration_range
    ORDER BY duration_range
""")
```

```python
x = np.arange(len(durationwise_credibilty))  # 0,1,2,...
width = 0.35

# Bar chart
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, durationwise_credibilty['good_credit'], width, label='Good Credit', color='green',alpha = 0.9)
plt.bar(x + width/2, durationwise_credibilty['bad_credit'], width, label='Bad Credit', color='red',alpha = 0.9)
plt.plot(durationwise_credibilty['duration_range'],durationwise_credibilty['bad_rate'],linewidth=2.5, marker = 'o',color = 'black', label = "Bad Rate(%)")
plt.legend()
plt.xticks(ticks=x, labels=durationwise_credibilty['duration_range'], rotation=45)
plt.show()
```

```python
agewise = run_query("""
    SELECT
        CASE
            WHEN Age_years < 25 THEN '1 Young People'
            WHEN Age_years < 35 THEN '2 Emerging Professionals'
            WHEN Age_years < 45 THEN '3 Adults'
            WHEN Age_years < 55 THEN '4 Pre-Retirement Indiviuals'
            ELSE '5 Senior cetizens'
        END as age_group,
        COUNT(*) as total,
        SUM(Creditability) as good_credit,
        COUNT(*) - SUM(Creditability) as bad_credit,
        ROUND((COUNT(*) - SUM(Creditability)) * 100.0 / COUNT(*), 2) as bad_rate,
        ROUND(AVG(Credit_Amount), 2) as avg_credit_amount,
        COUNT(Credit_Amount) as Number_of_customers
    FROM credit_data
    GROUP BY age_group
    ORDER BY age_group
""")
print(agewise)
x = np.arange(len(agewise))
width = 0.35

# Bar chart
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, agewise['good_credit'], width, label='Good Credit', color='green',alpha = 0.9)
plt.bar(x + width/2, agewise['bad_credit'], width, label='Bad Credit', color='red',alpha = 0.9)
plt.plot(agewise['age_group'],agewise['bad_rate'],linewidth=2.5, marker = 'o',color = 'black', label = "Bad Rate(%)")

plt.legend()
plt.xticks(ticks=x, labels=agewise['age_group'], rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 4))
fig, ax = plt.subplots(1,2,figsize=(14,4))
ax[0].plot(agewise['age_group'],agewise['Number_of_customers'],linewidth=2.5, marker = 'o',color = 'black', label = "Number of Customers")
ax[1].plot(agewise['age_group'],agewise['avg_credit_amount'],linewidth=2.5, marker = 'o',color = 'black', label = "Average Credit Amount")

plt.show()
```

```python
features = ['Age_years','Duration_of_Credit_monthly','Account_Balance','No_of_Credits_at_this_Bank','Sex_Marital_Status']
target = 'Creditability'
```

```python
# Shuffle the credit_data DataFrame while keeping it as a DataFrame
df_shuffled = credit_data.sample(frac=1, random_state=42).reset_index(drop=True)

X = df_shuffled[features]
y = df_shuffled[target]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Initialize and train the Random Forest classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
rf.fit(X_train, y_train)

# Make predictions
y_pred = rf.predict(X_test)

# Evaluate the model
print("\nClassification Report:")
print(classification_report(y_test, y_pred)) # Removed target_names as le_target is not defined

# Confusion matrix
print("\nConfusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Not Credible', 'Credible'], # Added explicit labels
            yticklabels=['Not Credible', 'Credible']) # Added explicit labels
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Feature importance
feature_importance = pd.DataFrame({
    'Feature': features,
    'Importance': rf.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nFeature Importance:")
print(feature_importance)

# Plot feature importance
plt.figure(figsize=(10, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance)
plt.title('Feature Importance')
plt.show()
```

# Data Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

credit_data = pd.read_csv('/content/Loan_default[1].csv')
```

# SQL - Connect

```python
# Create an in-memory SQLite database
df = credit_data
import sqlite3
conn = sqlite3.connect(':memory:')
df.to_sql('credit_data', conn, index=False, if_exists='replace')
def run_query(query):
    return pd.read_sql(query, conn)
run_query("SELECT * FROM credit_data LIMIT 5")
```

1. AgeGroup


```python
age_wise = run_query("""
   with age_group as (
       select *, case
           when "Age" < 18 then '0 minor'
           when "Age" >= 18 and "Age" < 30 then '18-20'
           when "Age" >= 20 and "Age" < 30 then '20-30'
           when "Age" >= 30 and "Age" < 40 then '30-40'
           when "Age" >= 40 and "Age" < 50 then '40-50'
           when "Age" >= 50 and "Age" < 60 then '50-60'
           else '60+'
       end as "AgeGroup"
       from credit_data
   )
   select
       "AgeGroup",
       count(*) as total_count,
       sum(case when "Default" = 1 then 1 else 0 end) as default_count,
       sum(case when "Default" = 0 then 1 else 0 end) as non_default_count,
       sum(case when "Default" = 1 then 1 else 0 end) * 100.0 / count(*) as default_percentage
   from age_group
   group by "AgeGroup"
   order by "AgeGroup"
""")


age_wise
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(len(age_wise))

fig, ax1 = plt.subplots(figsize=(10, 6))

# --- Bar chart: total credits ---
ax1.bar(x, age_wise['total_count'], color='skyblue', alpha=0.8, label="Total Credits")
ax1.set_xlabel("Age Group")
ax1.set_ylabel("Total Credits")
ax1.set_xticks(x)
ax1.set_xticklabels(age_wise['AgeGroup'], rotation=45)

# --- Line chart: default % ---
ax2 = ax1.twinx()
ax2.plot(x, age_wise['default_percentage'] * 100, marker='o', color='darkred',
         linewidth=2.5, label="Default %")
ax2.set_ylabel("Default %")

# --- Legends ---
bars_legend, labels1 = ax1.get_legend_handles_labels()
line_legend, labels2 = ax2.get_legend_handles_labels()
ax1.legend(bars_legend + line_legend, labels1 + labels2, loc="upper right")

plt.title("Age Group: Total Credits vs Default Percentage")
plt.tight_layout()
plt.show()
```

# Encoding - Feature Engineering

```python
df = credit_data.copy()
```

1. Education

```python
# mapping done in accordance to EDA
maping_of_education={'High School':0,"Bachelor's":1,"Master's":2,'PhD':3}
df['Education']=df['Education'].map(maping_of_education)
df.head()
```

2. Mortgage, CoSigner, Dependents Status

```python
# 'Yes': 1, 'No': 0
# For Yes there default rates are lower compared to 'No'
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['HasMortgage']=le.fit_transform(df['HasMortgage'])
df['HasCoSigner']=le.fit_transform(df['HasCoSigner'])
df['HasDependents']=le.fit_transform(df['HasDependents'])
df.head()
```

3. Employment Status

```python
maping_of_employment_type={'Unemployed':0,'Part-time':1,'Self-employed':2,'Full-time':3}
df['EmploymentType']=df['EmploymentType'].map(maping_of_employment_type)
df.head()
```

4. Marital Status

```python
maping_of_MaritalStatus={'Divorced':0,'Single':1,'Married':2}
df['MaritalStatus']=df['MaritalStatus'].map(maping_of_MaritalStatus)
df.head()
```

5. Loan Purpose

we'll use the One Hot Encoder for loan_purpose since it is a nominal data

```python
enconde_loan_purpose=pd.get_dummies(df['LoanPurpose'],drop_first=True,dtype=int)
df=pd.concat([df,enconde_loan_purpose],axis=1)
df.drop('LoanPurpose',axis=1,inplace=True)
df.head()
```
